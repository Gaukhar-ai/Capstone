```python
import datetime as dt
import scipy.stats as stats
import matplotlib.pyplot as plt
from matplotlib import style
import pandas as pd
import pandas_datareader.data as web
from scipy.stats import skew
from scipy.stats import kurtosis
import seaborn as sns
import numpy as np
```

```python
energy_df = pd.read_csv('energy_data.csv')
fin_df = pd.read_csv('fin_data.csv')
mater_df = pd.read_csv('mater_data.csv')
cons_stap_df = pd.read_csv('cons_stap_data.csv')
indust_df = pd.read_csv('indust_data.csv')
comms_df = pd.read_csv('comms_data.csv')
real_df = pd.read_csv('real_data.csv')
tech_df = pd.read_csv('tech_data.csv')
util_df = pd.read_csv('util_data.csv')
cons_disc_df = pd.read_csv('cons_disc_data.csv')
health_df = pd.read_csv('health_data.csv')

energy_df['sector'] = 'ENERGY'
energy_df['returns'] = energy_df['close'].pct_change()
energy_df['returns'] = energy_df['returns'] * 100
energy_df.volume = energy_df.volume.astype(float)

fin_df['sector'] = 'FIN'
fin_df['returns'] = fin_df['close'].pct_change()
fin_df['returns'] = fin_df['returns'] * 100
fin_df.volume = fin_df.volume.astype(float)

mater_df['sector'] = 'MATER'
mater_df['returns'] = mater_df['close'].pct_change()
mater_df['returns'] = mater_df['returns'] * 100
mater_df.volume = mater_df.volume.astype(float)

cons_stap_df['sector'] = 'CONS STAP'
cons_stap_df['returns'] = cons_stap_df['close'].pct_change()
cons_stap_df['returns'] = cons_stap_df['returns'] * 100
cons_stap_df.volume = cons_stap_df.volume.astype(float)

indust_df['sector'] = 'INDUST'
indust_df['returns'] = indust_df['close'].pct_change()
indust_df['returns'] = indust_df['returns'] * 100
indust_df.volume = indust_df.volume.astype(float)

comms_df['sector'] = 'COMMS'
comms_df['returns'] = comms_df['close'].pct_change()
comms_df['returns'] = comms_df['returns'] * 100
comms_df.volume = comms_df.volume.astype(float)

real_df['sector'] = 'REAL'
real_df['returns'] = real_df['close'].pct_change()
real_df['returns'] = real_df['returns'] * 100
real_df.volume = real_df.volume.astype(float)

tech_df['sector'] = 'TECH'
tech_df['returns'] = tech_df['close'].pct_change()
tech_df['returns'] = tech_df['returns'] * 100
tech_df.volume = tech_df.volume.astype(float)

util_df['sector'] = 'UTIL'
util_df['returns'] = util_df['close'].pct_change()
util_df['returns'] = util_df['returns'] * 100
util_df.volume = util_df.volume.astype(float)

cons_disc_df['sector'] = 'CONS DISC'
cons_disc_df['returns'] = cons_disc_df['close'].pct_change()
cons_disc_df['returns'] = cons_disc_df['returns'] * 100
cons_disc_df.volume = cons_disc_df.volume.astype(float)

health_df['sector'] = 'HEALTH'
health_df['returns'] = health_df['close'].pct_change()
health_df['returns'] = health_df['returns'] * 100
health_df.volume = health_df.volume.astype(float)

stocks_df = pd.concat([energy_df, fin_df, mater_df, cons_stap_df, indust_df, comms_df, real_df, tech_df, util_df, cons_disc_df, health_df], sort=False)

#Set index
stocks_df.set_index(['date'], inplace=True)

#Print DataFrame
stocks_df
```

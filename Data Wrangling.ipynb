Next, I save each sector DataFrame locally as a csv file for easy retrieval, concatenate all DataFrames to define stocks_df, and add the 'returns' column. Finally, I set a hierarchical index and display the DataFrame.

```python
#Save each DataFrame locally as a csv file
energy_df = pd.read_csv('energy_data.csv')
fin_df = pd.read_csv('fin_data.csv')
mater_df = pd.read_csv('mater_data.csv')
cons_stap_df = pd.read_csv('cons_stap_data.csv')
indust_df = pd.read_csv('indust_data.csv')
comms_df = pd.read_csv('comms_data.csv')
real_df = pd.read_csv('real_data.csv')
tech_df = pd.read_csv('tech_data.csv')
util_df = pd.read_csv('util_data.csv')
cons_disc_df = pd.read_csv('cons_disc_data.csv')
health_df = pd.read_csv('health_data.csv')

#Concatenate all DataFrames into one large DataFrame
stocks_df = pd.concat([energy_df, fin_df, mater_df, cons_stap_df, indust_df, comms_df, real_df, tech_df, util_df, cons_disc_df, health_df], sort=False)

#Add 'returns' column
stocks_df['returns'] = stocks_df['close'].pct_change()
stocks_df['returns'] = stocks_df['returns'] * 100
stocks_df.volume = stocks_df.volume.astype(float)

#Set hierarchical index
stocks_df.set_index(['symbol', 'date'], inplace=True)
stocks_df.sort_index()

#Print DataFrame
stocks_df
```

```python
#Pivot method to create DataFrames for each sector with each symbol as a column and their close prices as the values
energy_p = energy_df.pivot(index='date', columns= 'symbol', values='close')
mater_p = mater_df.pivot(index='date', columns= 'symbol', values='close')
fin_p = fin_df.pivot(index='date', columns= 'symbol', values='close')
cons_stap_p = cons_stap_df.pivot(index='date', columns= 'symbol', values='close')
indust_p = indust_df.pivot(index='date', columns= 'symbol', values='close')
comms_p = comms_df.pivot(index='date', columns= 'symbol', values='close')
real_p = real_df.pivot(index='date', columns= 'symbol', values='close')
tech_p = tech_df.pivot(index='date', columns= 'symbol', values='close')
util_p = util_df.pivot(index='date', columns= 'symbol', values='close')
cons_disc_p = cons_disc_df.pivot(index='date', columns= 'symbol', values='close')
health_p = health_df.pivot(index='date', columns= 'symbol', values='close')
```

```python
#Find stocks with missing data
sectors = [energy_p, fin_p, mater_p, cons_stap_p, indust_p, comms_p, real_p, tech_p, util_p, cons_disc_p, health_p]

for sector in sectors:
    print(sector.describe())
```

```python
#Drop stocks with missing data from price datasets
cons_stap_p.drop('KHC', axis=1, inplace=True)
comms_p.drop('TWTR', axis=1, inplace=True)
real_p.drop('CBRE', axis=1, inplace=True)
mater_p.drop('DWDP', axis=1, inplace=True)
mater_p.drop('WRK', axis=1, inplace=True)
```

```python
#Concatenate all sectors 
stock_p = pd.concat([energy_p, fin_p, mater_p, cons_stap_p, indust_p, comms_p, real_p, tech_p, util_p, cons_disc_p, health_p], axis=1)
```

```python
#Add returns column to each sector DataFrame
sectors_r = [energy_df, fin_df, mater_df, cons_stap_df, indust_df, comms_df, real_df, tech_df, util_df, cons_disc_df, health_df]

for sector in sectors_r:
    sector['returns'] = sector['close'].pct_change()
    sector['returns'] = sector['returns'] * 100

#Pivot method to create DataFrames for each sector with each symbol as a column and their returns as the values
energy_r = energy_df.pivot(index='date', columns= 'symbol', values='returns')
mater_r = mater_df.pivot(index='date', columns= 'symbol', values='returns')
fin_r = fin_df.pivot(index='date', columns= 'symbol', values='returns')
cons_stap_r = cons_stap_df.pivot(index='date', columns= 'symbol', values='returns')
indust_r = indust_df.pivot(index='date', columns= 'symbol', values='returns')
comms_r = comms_df.pivot(index='date', columns= 'symbol', values='returns')
real_r = real_df.pivot(index='date', columns= 'symbol', values='returns')
tech_r = tech_df.pivot(index='date', columns= 'symbol', values='returns')
util_r = util_df.pivot(index='date', columns= 'symbol', values='returns')
cons_disc_r = cons_disc_df.pivot(index='date', columns= 'symbol', values='returns')
health_r = health_df.pivot(index='date', columns= 'symbol', values='returns')

#Set first row equal to NaN
energy_r.iloc[0,:] = np.nan
mater_r.iloc[0,:] = np.nan
fin_r.iloc[0,:] = np.nan
cons_stap_r.iloc[0,:] = np.nan
indust_r.iloc[0,:] = np.nan
comms_r.iloc[0,:] = np.nan
real_r.iloc[0,:] = np.nan
tech_r.iloc[0,:] = np.nan
util_r.iloc[0,:] = np.nan
cons_disc_r.iloc[0,:] = np.nan
health_r.iloc[0,:] = np.nan


#Drop stocks with missing data from returns datasets
cons_stap_r.drop('KHC', axis=1, inplace=True)
comms_r.drop('TWTR', axis=1, inplace=True)
real_r.drop('CBRE', axis=1, inplace=True)
mater_r.drop('DWDP', axis=1, inplace=True)
mater_r.drop('WRK', axis=1, inplace=True)


#Concatenate all sectors
stock_r = pd.concat([energy_r, fin_r, mater_r, cons_stap_r, indust_r, comms_r, real_r, tech_r, util_r, cons_disc_r, health_r], axis=1)
```

```python
#Save each DataFrame to csv for easy retrival
energy_r.to_csv('energy_returns.csv')
mater_r.to_csv('mater_returns.csv')
fin_r.to_csv('fin_returns.csv')
cons_stap_r.to_csv('cons_stap_returns.csv')
indust_r.to_csv('indust_returns.csv')
comms_r.to_csv('comms_returns.csv')
real_r.to_csv('real_returns.csv')
tech_r.to_csv('tech_returns.csv')
util_r.to_csv('util_returns.csv')
cons_disc_r.to_csv('cons_disc_returns.csv')
health_r.to_csv('health_returns.csv')

energy_p.to_csv('energy_prices.csv')
mater_p.to_csv('mater_prices.csv')
fin_p.to_csv('fin_prices.csv')
cons_stap_p.to_csv('cons_stap_prices.csv')
indust_p.to_csv('indust_prices.csv')
comms_p.to_csv('comms_prices.csv')
real_p.to_csv('real_prices.csv')
tech_p.to_csv('tech_prices.csv')
util_p.to_csv('util_prices.csv')
cons_disc_p.to_csv('cons_disc_prices.csv')
health_p.to_csv('health_prices.csv')

stock_r.to_csv('stock_returns.csv')
stock_p.to_csv('stock_prices.csv')
```
